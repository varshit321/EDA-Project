```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('jr2.csv')
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
jr2=df
```

```python
jr2
```

### Histogram

##### Prices of bustickets of HYD TO BLR

```python
plt.figure(figsize=(8,3))
plt.hist(df['prices'],bins=10,color='purple')
plt.xlabel("Prices")
plt.ylabel("Frequency")
plt.title("Histogram of prices")
plt.show()
```

##### Ratings of bustickets from HYD TO BLR

```python
plt.figure(figsize=(6,3))
plt.hist(df['ratings'],bins=10,color='red')
plt.xlabel("Ratings")
plt.ylabel("Frequency")
plt.title("Histogram of ratings")
plt.show()
```

##### Prices of bustickets based on AC_type

```python
plt.style.use('dark_background')
sns.histplot(data=df,x='prices',bins=20,color='red',hue='AC_type').set(title='Prices of bustickets based on AC_type')
plt.show()
```

##### Prices of bustickets based on sleeper

```python
sns.set_style('dark')
sns.histplot(data=df,x='prices',bins=20,color='red',hue='sleeper').set(title='Prices of bustickets based on sleeper')
plt.show()
```

##### Prices of bustickets based on seater

```python
plt.style.use('classic')
sns.histplot(data=jr2,x='prices',bins=20,color='red',hue='Seater').set(title='Prices of bustickets based on seater')
plt.show()
```

##### Prices of bustickets based on semi_seater

```python
sns.set_style('darkgrid')
sns.histplot(data=jr2,x='prices',bins=20,color='purple',hue='semi_sleeper').set(title='Prices of bustickets based on semi_seater')
plt.show()
```

##### Prices of bustickets based on seater

```python
plt.style.use('dark_background')
sns.histplot(data=jr2,x='ratings',bins=20,color='red',hue='Seater').set(title='Prices of bustickets based on seater')
plt.show()
```

## Categorical vs Categorical

##### Top 5 AC travel bus names

```python
pd.crosstab(jr2['Bus_Name'],jr2['AC_type']).sort_values('A/C',ascending=False).head()
```

##### Top 5 NON AC travel bus names

```python
pd.crosstab(jr2['Bus_Name'],jr2['AC_type']).sort_values('NON A/C',ascending=False).head()
```

##### A complete data of number of AC type and Seater type buses

```python
s1=pd.crosstab(df['sleeper'],df['AC_type'])
s2=pd.crosstab(df['semi_sleeper'],df['AC_type'])
s3=pd.crosstab(df['Seater'],df['AC_type'])
s=pd.concat([s1,s2,s3],axis=0)
s.index = ['non sleeper', 'sleeper', 'non semi_sleeper', 'semi_sleeper', 'non seater', 'seater']
```
